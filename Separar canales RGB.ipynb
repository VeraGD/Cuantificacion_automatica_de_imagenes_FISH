# Obtener im√°genes de los distintos canales rgb de la imagen a color

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

from skimage import io
import imageio
```

## Cargar imagen original

```python
imagen = io.imread(fname="imageFISH.jpg")
```

```python
io.imshow(imagen)
```

## Funciones para crear las tres im√°genes de los tres canales

```python
def canal_r(image):
    f,c,a = image.shape
    canal = np.zeros((f,c))
    for i in range(0,f):
        for j in range(0,c):
            canal[i][j] = image[i][j][0]
    return canal
```

```python
imagen_r = canal_r(imagen)
```

```python
def canal_g(image):
    f,c,a = image.shape
    canal = np.zeros((f,c))
    for i in range(0,f):
        for j in range(0,c):
            canal[i][j] = image[i][j][1]
    return canal
```

```python
imagen_g = canal_g(imagen)
```

```python
def canal_b(image):
    f,c,a = image.shape
    canal = np.zeros((f,c))
    for i in range(0,f):
        for j in range(0,c):
            canal[i][j] = image[i][j][2]
    return canal
```

```python
imagen_b = canal_b(imagen)
```
